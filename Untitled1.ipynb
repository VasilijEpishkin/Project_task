```python
from Bio import Entrez
import pandas as pd
from Bio import SeqIO

# Устанавливаем email для использования в качестве идентификатора
Entrez.email = "vasa.epishkin990@gmail.com"

# Загрузим файл в DataFrame
df = pd.read_csv('C:/Users/vasae/PycharmProjects/pythonProject1/reac_data.csv', encoding='unicode_escape')

# Получим список первых 10 значений из колонки Protein
protein_values = df["Protein"].tolist()

# Создаем пустой список для результатов запросов
results = []

# Проходим по всем значениям в списке protein_values
for protein in protein_values:
    # Формируем запрос к NCBI
    query = f"{protein}[Protein]"
    handle = Entrez.esearch(db="protein", term=query)
    record = Entrez.read(handle)

    # Получаем первый id из списка результатов
    if record['Count'] != '0':
        uniprot_id = record['IdList'][0]
        handle = Entrez.efetch(db='protein', id=uniprot_id, rettype='fasta', retmode='text')
        seq_record = SeqIO.read(handle, 'fasta')
        handle.close()
        seq = seq_record.seq
        results.append(seq)

        # Печатаем результат запроса
        print(f"Запрос {protein} успешно выполнен. Результат: {seq}")
    else:
        results.append(None)

        # Печатаем сообщение об ошибке
        print(f"Запрос {protein} не дал результатов")

# Создаем новый DataFrame с результатами запросов
result_df = pd.DataFrame({"Protein": protein_values, "Sequence": results})

# Записываем результаты в новый CSV файл
result_df.to_csv('results.csv', index=False)
```
